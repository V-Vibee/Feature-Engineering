### Black Friday EDA & Feature engineering

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Importing the dataset

df_train= pd.read_csv('train.csv')
df_train.head()
```

```python
df_train.shape
```

```python
## Import the test data

df_test= pd.read_csv('test.csv')
df_test.head()
```

```python
df_test.shape
```

```python
# Merging  both dataset

df= pd.concat((df_train, df_test), ignore_index=True)
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
# Dropping the Userid column( axis=1 means ROW-WISE, Axis 0 means COLUMN -WISE)

df.drop(['User_ID'], axis=1, inplace= True)
```

```python
df.head()
```

```python
# Handling category features Gender, Map function changes the value easily.....

df['Gender']= df['Gender'].map({'F':0, 'M':1})
```

```python
df.head()
```

```python
# Handle Categorical feature Age

df['Age'].unique()
```

```python
# Target Ordinal encoding

df['Age']= df['Age'].map({'0-17':1, '18-25':2, '26-35':3, '36-45':4, '46-50':5, '51-55':6, '55+':7})
```

```python
df.head()
```

```python
# Fixing Categorical field in City_Category column

df_city = pd.get_dummies(df['City_Category'], prefix='City')

df_city = df_city.astype(int)
```

```python

df_city.head()
```

```python
df=pd.concat([df, df_city], axis=1)
```

```python
df.head()
```

```python
df.drop('City_Category', axis=1, inplace= True)
```

```python
df.head()
```

```python
df.head()
```
